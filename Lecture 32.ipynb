```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits import mplot3d
```

```python
x=np.linspace(start=-1,stop=1,num=1000)
y=x
X,Y= np.meshgrid(x,y)
Z=X**2+Y**2
# Z_=-np.sqrt(X**2+Y**2)
# fig = plt.figure(figsize=(4,7))
ax=plt.axes()
ax.contour(X, Y,Z, 50)
# ax.plot_surface(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')


# ax.scatter(initial_points[0],initial_points[1],initial_points[2],color="red")
def gradient(points):
    x=points[0]
    y=points[1]
    return np.array([2*x,2*y])
## perform gradient descent 
initial_points=np.array([[0.1],[1]])
points=[initial_points]
for i in range(100):
    if gradient(initial_points)[0]**2 +gradient(initial_points)[1]**2<10e-7:
        break
    initial_points=initial_points-0.1*gradient(initial_points)
    # print(initial_points)
    points.append(initial_points)
points=np.array(points)
# print(points)
ax.plot(points[:,0],points[:,1],color="red")
plt.show()
```

```python
x=np.linspace(start=-1,stop=1,num=1000)
y=x
X,Y= np.meshgrid(x,y)
Z=X**2+100*(Y**2)
# Z_=-np.sqrt(X*2+Y*2)
# fig = plt.figure(figsize=(4,7))
ax=plt.axes()
ax.contour(X, Y,Z, 50)
# ax.plot_surface(X, Y, Z)
ax.set_xlabel('x')
ax.set_ylabel('y')

def val(points):
    return points[0]**2+100*points[1]**2
# ax.scatter(initial_points[0],initial_points[1],initial_points[2],color="red")
def gradient(points):
    x=points[0]
    y=points[1]
    return np.array([2*x,200*y])
## perform gradient descent 
initial_points=np.array([[1],[1]])
points=[initial_points]
step_size=0.01
alpha=0.01
old_point=initial_points
# print(old_point)
for i in range(100):
    for j in range(100):
        # print(initial_points)
        # print(gradient(initial_points))
        new_point=initial_points-alpha*gradient(initial_points)
        print("{iteration}:-{old}:{old_val}      {new}:{new_val}".format(iteration=i,old=old_point,old_val=val(old_point),new=new_point,new_val=val(new_point)))
        if val(new_point)>val(old_point):
            points.append(old_point)
            initial_points=old_point
            alpha=0.01
            break
        else:
            alpha+=step_size
            old_point=new_point
points=np.array(points)
# print(points)
ax.plot(points[:,0],points[:,1],color="red")
# print(points)
plt.show()
```
