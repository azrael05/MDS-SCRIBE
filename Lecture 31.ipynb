```python
import matplotlib.pyplot as plt
import numpy as np
```

## Z<sup>2</sup> = X<sup>2</sup> + Y<sup>2</sup> Function Surface Plot

```python
x=np.linspace(start=-1,stop=1,num=1000)    ## Gnerating 1000 points between -1 and 1
y=x
X,Y= np.meshgrid(x,y)
Z=X**2+Y**2             ## Defining the function

## Creating Surface plot of the function
fig = plt.figure(figsize=(4,7))
ax=plt.axes(projection="3d")
ax.plot_surface(X, Y, Z)
ax.set_xlabel('X')
ax.set_ylabel('Y')

plt.show()
```

## Z<sup>2</sup> = X<sup>2</sup> + Y<sup>2</sup> Function Contour Plot with Tangent Plane

```python
## Gnerating Contour plot of the function
x=np.linspace(start=-3,stop=3,num=1000)
y=x
X,Y= np.meshgrid(x,y)
Z=X**2+Y**2
Z_TAN=2*X+4*Y-5
fig = plt.figure
ax=plt.axes(projection="3d")
ax.contour3D(X, Y, Z,40)

## Plotting the point of contact/intersection
ax.scatter([1], [2], [5], color="red")


## Generating the Contour of Tangent plane
X=np.linspace(-2,3,100)
Y=np.linspace(0,3,100)
X,Y=np.meshgrid(X,Y)
Z_TAN=2*X+4*Y-[5]
ax.contour3D(X,Y,Z_TAN,50,cmap="binary")
ax.set_xlabel('x')
ax.set_ylabel('y')
plt.show()
```

## Generating 2D contour of  Function- Z<sup>2</sup> = X<sup>2</sup> + Y<sup>2</sup>

```python
x=np.linspace(start=-1,stop=1,num=1000)
y=x
X,Y= np.meshgrid(x,y)
Z=X**2+Y**2
fig = plt.figure()
ax=plt.axes()
ax.contour(X, Y, Z , 50)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_yticks([-1,-0.5,0,0.5,1])
ax.set_xticks([-1,-0.5,0,0.5,1])
ax.set_aspect('equal')
plt.show()
```

## Generating 2D contour of  Function: Z<sup>2</sup> = X<sup>2</sup> + Y<sup>2</sup> along with the tangent line and point of contact

```python
fig = plt.figure
ax=plt.axes()

## Generating the Contour of Function
x=np.linspace(start=-3,stop=3,num=1000)
y=x
X,Y= np.meshgrid(x,y)
Z=X**2+Y**2
ax.contour(X, Y, Z,40)


## Plotting the point of contact/intersection
ax.scatter([1],[2],[5],color="red")


## Generating the Tangent line while keeping Z fixed at 5 
x=np.linspace(-1,3,1000)
y=(10-2*x)/4    ## Equation of the tangent plane z=2x-4y+5 ==> @z=5, y=(10-2x)/4
ax.plot(x,y)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_aspect('equal')
plt.show()
```

## Example 2 of Contour Plane f(x,y,z)=x<sup>2</sup>+y<sup>2</sup>+z<sup>2</sup>

```python
## Setting colour codes
color=["red","black","yellow","purple"]
fig = plt.figure()
ax=plt.axes(projection="3d")
    
for f_value in range(1,4):          ## Varying the function value between 1 and 3
    
    x=np.linspace(start=-f_value,stop=f_value,num=1000)
    y=x
    X,Y= np.meshgrid(x,y)
    Z=f_value**2-X**2-Y**2
    for x_ in range(Z.shape[0]):
        for y_ in range(Z.shape[1]):
            Z[x_][y_]=np.sqrt(Z[x_][y_] if Z[x_][y_]>0 else 0)   ## For those values where Z^2<0 i.e. function does't exist, we set Z=0
    
    # Combining the two halfs of the sphere
    Z=np.concatenate([Z,-Z])
    Y=np.concatenate([Y,Y])
    X=np.concatenate([X,X])

    # Plotting Contour
    ax.contour3D(X, Y, Z, 10,colors=color[f_value-2])

ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.xlim(-5,5)
plt.ylim(-5,5)
plt.show()
```
